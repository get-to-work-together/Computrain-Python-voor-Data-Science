# Demo lineaire regressie

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 1000
```

```python
x = np.linspace(1, 100, n)
```

```python
y_calculated = 500 + 0.5 * x
y = y_calculated + np.random.randn(n) * 60
```

```python
plt.scatter(x, y, alpha=0.5)
plt.show()
```

```python
np.corrcoef(x, y)
```

```python
fit = np.polynomial.Polynomial.fit(x, y, deg=1).convert()
fit
```

```python
type(fit)
```

```python
fit.coef
```

```python
a, b = fit.coef
# a, b, c, d, e, f, g, h = fit.coef
```

```python
regression_y = a + b * x # + c * x**2 + d * x**3 + e * x**4 + f * x**5 + g * x**6 + h * x**7
```

```python
plt.scatter(x, y, alpha=0.5)
plt.plot(x, y_calculated, color='yellow')
plt.plot(x, regression_y, color='red')
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({'x': x, 'y': y})

sns.lmplot(data=df, x='x', y='y', ci=95);
```

```python
scipy.stats.linregress(x, y)
```

```python
errors = y - regression_y
```

```python
plt.hist(errors, bins=20)
plt.show()
```

```python
mae = np.abs(errors).mean()
mae
```

```python
rmse = np.sqrt(np.mean(errors**2))
rmse
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y, regression_y)
```

```python
mean_squared_error(y, regression_y) ** 0.5
```

```python
r2_score(y, regression_y)
```
