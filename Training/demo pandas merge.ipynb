```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'key': ['A', 'A', 'B', 'C', 'E'], 'value': [1, 2, 3, 5, 9]})
df1
```

```python
df2 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value2': [5, 6, 7, 8]})
df2
```

```python
pd.merge(df1, df2, left_on='key', right_on='key', how='outer', suffixes=('_left', '_right'))
```

```python
df_concated = pd.concat([df1, df1])
df_concated
```

```python
df_concated.reset_index(drop=True)
```

```python
df_concated.set_index('key')
```
