```python
def create_voltage_figure(a=2.0, b=1.0, c=1.0):
    t = np.arange(0.0, 2.0, 0.01)
    s = c + b * np.sin(a * np.pi * t)

    fig, ax = plt.subplots(figsize=(4,3))

    ax.plot(t, s)

    ax.set(xlabel='time (s)', 
           ylabel='voltage (mV)',
           title='Voltage')
    
    ax.set_ylim(-6, 6)
    
    ax.grid()

    plt.close(fig) # CLOSE THE FIGURE!
    return fig

# pn.pane.Matplotlib(create_voltage_figure(), dpi=144, tight=True)

pn.interact(create_voltage_figure, a=10.0, b=1.0, c=5.0)
```
