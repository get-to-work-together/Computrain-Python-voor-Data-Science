```python
df['make'] = df['make'].replace({
    'chevroelt': 'chevrolet',
    'chevy': 'chevrolet',
    'maxda': 'mazda',
    'mercedes-benz': 'mercedes',
    'toyouta': 'toyota',
    'vokswagen': 'volkswagen',
    'vw': 'volkswagen'
})
```

```python
bins = [0, 2000, 4000, 6000]
labels = ['light', 'medium', 'heavy']
df['weight_class'] = pd.cut(df['weight'], bins, labels = labels)
```

```python
df['weight'].plot(kind='hist')
```

```python
df
```

```python
bins = df['weight'].quantile([0, 0.35, 0.7, 1.0]).values
df['weight_class'] = pd.cut(df['weight'], bins, labels = labels)
```

```python
df['weight'].hist(bins=20)
```

```python
df['weight_class'].value_counts().plot(kind='bar')
```

```python
df.pivot_table(index = 'origin', 
               columns = 'weight_class', 
               aggfunc = 'count',
               values = 'mpg',
               fill_value = 0,
               observed = False)
```
